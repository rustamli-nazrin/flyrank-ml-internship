## 1. My lane as an ML task (type)

I am focusing on the Refresh / Content Opportunity Scoring lane. This is a Scoring and Ranking task.My goal is to calculate a priority score for web pages and rank them so a human reviewer can easily see which pages need attention first.

```python
selected_lane = "Refresh / Content Opportunity Scoring"
task_type = "Scoring / Ranking"
print(f"Lane: {selected_lane} | Task Type: {task_type}")
```

## 2. Target or proxy

​The target is a binary proxy label indicating whether a content page is experiencing a drop in search performance. We determine this using the observed signal trend_direction == "down" from our dataset.

```python
def get_target_label(dataframe):
    return dataframe['trend_direction'] == 'down'
```

##3. Success metric

The primary success metric will be Precision@50. This metric measures how many of the top 50 highly ranked pages flagged by the system are actually experiencing a real decline. This aligns perfectly with real-world operational capacity, ensuring human editors do not waste time on false positives.

```python
target_precision_at_50 = 0.70
print(f"Goal metric: Precision@50 >= {target_precision_at_50}")

```

## 4. The unit of analysis, as a real dataframe

The unit of analysis (the data grain) is one unique pseudonymized content item (page) per client.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/rustamli-nazrin/flyrank-ml-internship/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

# Apply initial filters
df_clean = df[(df['impressions_90d'] > 0) & (df['content_age_days'] >= 90)]

# Deduplicate by content_id to establish our true grain
df_clean = df_clean.drop_duplicates(subset=['content_id'])

# Display the true unit of analysis
print("DataFrame Shape (Rows, Columns):", df_clean.shape)
df_clean[['content_id', 'client_id', 'impressions_90d', 'trend_direction']].head()
```

## 5. Why ML beats a fixed rule here

A static if-else rule creates a massive, noisy list that ignores overall search demand or user engagement patterns. Traffic shifts are caused by complex interactions between multiple signals like impression metrics, historic positions, and seasonal shifts. A machine learning model handles these non-linear relationships efficiently, separating actual traffic decay from minor data noise.

```python
all_rows = len(df_clean)
stale_pages_rule = len(df_clean[df_clean['content_age_days'] >= 180])

print(f"Total clean rows: {all_rows}")
print(f"Rows flagged by a simple age rule (>= 180 days): {stale_pages_rule} ({stale_pages_rule/all_rows:.1%})")
print("An ML system is needed to filter this down to the highest priority candidates.")
```
