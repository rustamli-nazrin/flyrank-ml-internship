```python
import pandas as pd

# Load dataset
df = pd.read_csv('../../data/raw/content_refresh_anonymized.csv')

total_items = len(df['content_id'].unique())

declining_count = len(df[df['trend_direction'] == 'down'])
pct_declining = (declining_count / len(df)) * 100

avg_age_days = df['content_age_days'].mean()

print(f"Real Number 1 (Total Content Items): {total_items:,}")
print(f"Real Number 2 (Percentage of active declining pages): {pct_declining:.2f}%")
print(f"Real Number 3 (Average content age across inventory): {avg_age_days:.1f} days")
```
